```python
plt.plot(x,y)
plt.title('Plot-1')
plt.xlabel('X-array')
plt.ylabel('Y-array')
```

```python
plt.plot(x,y, color = 'r', ls = ':', lw = '3', marker = 'o', markerfacecolor = 'y', markersize = 15, markeredgewidth = 2, markeredgecolor = 'g')
plt.title('Plot-1')
plt.xlabel('X-array')
plt.ylabel('y-array')
```

```python
plt.plot(x,y, color = 'r', ls = '--', lw = '3', marker = '*', markerfacecolor = 'y', markersize = 15, markeredgewidth = 2, markeredgecolor = 'g')
plt.title('Plot-1', fontdict = {'font':'monospace', 'color':'red', 'size' :25})
plt.xlabel('X-array', fontdict = {'font':'monospace', 'color':'red', 'size' :18})
plt.ylabel('y-array', fontdict = {'font':'monospace', 'color':'red', 'size' :18})
```

```python
plt.plot(x,y, label = 'X vs Y', color = 'y', ls = ':', lw = '3', marker = '@', markerfacecolor = 'y', markersize = 15, markeredgewidth = 2, markeredgecolor = 'b')
plt.title('Plot-1', fontdict = {'font':'monospace', 'color':'yellow', 'size' :25})
plt.Xlabel('x-array', fontdict = {'font':'monospace', 'color':'yellow', 'size' :18})
plt.ylabel('y-array', fontdict ={'font':'monospace', 'color':'red', 'size' :18})

plt.legend(loc = 4, fontsize = 30, labalcolor = 'r')
plt.show()
```
