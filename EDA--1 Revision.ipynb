```python
plt.figure(figsize=(20,10))

# Melt the dataframe to convert it to long-form
df_melted = pd.melt(df, id_vars=['BusinessTravel'], value_vars=['OverTime'])

sns.countplot(x='BusinessTravel', hue='value', data=df_melted)
```

```python
plt.figure(figsize=(20,10))

# Melt the dataframe to convert it to long-form
df_melted = pd.melt(df, id_vars=['BusinessTravel'], value_vars=['Gender'])

sns.countplot(x='BusinessTravel', hue='value', data=df_melted)
```

```python
income_group=df.groupby(by= 'JobRole').agg(Mean_Salary=('MonthlyIncome','mean'))
income_group
```

```python
income_group=df.groupby(by= 'JobRole').agg(Mean_Salary=('MonthlyIncome','median'))
income_group
```

```python
income_group.sort_values(by='Mean_Salary',ascending=False)
```

```python
income_group.sort_values(by='Mean_Salary',ascending=True)
```
