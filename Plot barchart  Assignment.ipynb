# Plot barchart and Piechart/donut chart for categorical variables

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
categories = ['Category A', 'Category B', 'Category C', 'category D']
counts = [25, 40, 15, 20]

df = pd.DataFrame({'Category': categories, 'Count': 'counts'})
```

```python
import seaborn as sns


plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Count', data=df, palette='viridis')
plt.title('Bar Chart of Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python

plt.figure(figsize=(8, 8))
plt.pie(df['Count'], labels=df['Category'], autopct='%1.1f%%', colors=sns.color_palette('viridis', len(categories)))
plt.title('Pie Chart of Categories')
plt.show()
```

```python

plt.figure(figsize=(8, 8))
plt.pie(df['Count'], labels=df['Category'], autopct='%1.1f%%', colors=sns.color_palette('viridis', len(categories)), wedgeprops=dict(width=0.4))
plt.title('Donut Chart of Categories')
plt.show()
```
