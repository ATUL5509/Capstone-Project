```python
# Group by on basis of Jobrole with avg monthly salary
```

```python
income_group=df.groupby(by='JobRole').agg(Mean_salary=('MonthlyIncome','mean'))
income_group
```

```python
income_group.sort_values(by='Mean_salary',ascending=False)
```

```python
plt.figure(figsize=(5,5))

sns.barplot(x=income_group.index,y=income_group.Mean_salary)
plt.xticks(rotation=90)
plt.show()
```
